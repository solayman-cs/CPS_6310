# Exercise 1: Fair Coin Simulation

```python
import random 
import matplotlib.pyplot as plt 

result = []
n = 0
for i in range(1, 101):
    x1 = random.randint(0,1)  # 0 for head and 1 for tail
    x2 = random.randint(0,1)

    if (x1==0 and x2 ==0):
        n = n + 1 
    result.append(n/i)


plt.title('trails: 100')
plt.plot(result)
plt.axhline(y=0.25, color='r', linestyle='--', label='Theoretical Probability (1/4)')
plt.show()
```

# Exercise 2: Chess Players Simulation

```python
import numpy as np
import matplotlib.pyplot as plt


def chess_tournament_simulation(total_games):
    players = {'A': 0, 'B': 0, 'C': 0}
    results = [('A', 'B', 250, 420), ('B', 'C', 150, 250), ('A', 'C', 98, 350)]
    
    for _ in range(total_games):
        match = results[np.random.choice(len(results))]  # Select a random match
        player1, player2, wins, total = match
        prob_win = wins / total
        
        if np.random.rand() < prob_win:
            players[player1] += 2  # Winner gets 2 points
        else:
            players[player2] += 2  # Loser gets 2 points

    return players

final_scores = chess_tournament_simulation(300)
print("Final Rankings:")
print(sorted(final_scores.items(), key=lambda x: x[1], reverse=True))
```

# Exercise 3: Carnival Game Simulation

```python
def carnival_game_simulation(trials):
    outcomes = np.random.randint(1, 11, size=trials)
    e1 = np.isin(outcomes, [2, 4, 6, 8, 10])
    e2 = np.isin(outcomes, [8, 9, 10])
    prob_one_prize = np.sum(e1 | e2) / trials
    return prob_one_prize

monte_carlo_probability = carnival_game_simulation(100000)
analytical_probability = 7 / 10  # P(E1 U E2)
prob_no_prize = 1 - analytical_probability

print(f"Monte Carlo estimated probability of winning at least one prize: {monte_carlo_probability}")
print(f"Analytical probability of winning at least one prize: {analytical_probability}")
print(f"Probability of not winning any prize analytically: {prob_no_prize}")
```
